```python
#Uploaded the dataset and explored the data
bankruptcy_data = pd.read_csv('/content/Taiwan_bankruptcy.csv')

print("Dataset shape:", bankruptcy_data.shape)
display(bankruptcy_data.head())
display(bankruptcy_data.info())
display(bankruptcy_data.describe())
```

```python
#Checked for missing values
missing = bankruptcy_data.isnull().sum()
total_missing = missing.sum()

print(f"\nTotal missing values in dataset: {total_missing}")
print("\nColumns with missing values:")
display(missing[missing > 0])

```

```python
sns.countplot(x='Bankrupt?', data=bankruptcy_data)
plt.title('Class Distribution (Bankrupt vs. Healthy)')
plt.show()

class_dist = bankruptcy_data['Bankrupt?'].value_counts(normalize=True) * 100
print("Class distribution (%):")
print(class_dist)
```

```python
print(bankruptcy_data.columns.tolist())
```

```python
bankruptcy_data.columns = bankruptcy_data.columns.str.strip()
```

```python
#CORRELATION HEATMAP
#Correlation matrix on selected financial ratios
corr = bankruptcy_data[[
    'Debt ratio %',
    'Working Capital to Total Assets',
    'Current Liability to Assets',
    'Liability-Assets Flag',
    'Net Income to Total Assets',
    'Net Income Flag'
]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Selected Features')
plt.tight_layout()
plt.show()
```

```python
#Correlation matrix (2) on updated financial ratios selection
corr = bankruptcy_data[[
    'Debt ratio %',
    'Working Capital to Total Assets',
    'Quick Ratio',
    'Liability-Assets Flag',
    'Net Income to Total Assets',
    'Net Income Flag'
]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Selected Features')
plt.tight_layout()
plt.show()
```

```python
selected_features = [
    'Debt ratio %',
    'Working Capital to Total Assets',
    'Quick Ratio',
    'Liability-Assets Flag',
    'Net Income to Total Assets',
    'Net Income Flag'
]
```

```python
#BOXPLOTS BY BANKRUPTCY CLASS
import matplotlib.pyplot as plt
import seaborn as sns

#Boxplot for each feature
for feature in selected_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Bankrupt?', y=feature, data=bankruptcy_data, palette='pastel')
    plt.title(f'{feature} by Bankruptcy Status')
    plt.xlabel('Bankrupt? (0 = No, 1 = Yes)')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

```python
#HISTOGRAMS OF FEATURE DISTRIBUTIONS
import matplotlib.pyplot as plt

#Histogram of feature distributions
selected_features = [
    'Debt ratio %',
    'Working Capital to Total Assets',
    'Quick Ratio',
    'Liability-Assets Flag',
    'Net Income to Total Assets',
    'Net Income Flag'
]

bankruptcy_data[selected_features].hist(bins=30, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Distributions of Key Financial Ratios")
plt.tight_layout()
plt.show()
```

```python
#SCATTERPLOTS BETWEEN SELECTED PAIRS
import matplotlib.pyplot as plt
import seaborn as sns

#Defined three key scatterplot pairs
scatter_pairs = [
    ('Debt ratio %', 'Net Income to Total Assets'),
    ('Working Capital to Total Assets', 'Net Income to Total Assets'),
    ('Debt ratio %', 'Working Capital to Total Assets')
]

#Plotted each scatterplot
for x_var, y_var in scatter_pairs:
    plt.figure(figsize=(7, 5))
    sns.scatterplot(
        x=x_var,
        y=y_var,
        hue='Bankrupt?',
        data=bankruptcy_data,
        palette={0: 'green', 1: 'red'},
        alpha=0.6
    )
    plt.title(f'{y_var} vs. {x_var} by Bankruptcy Status')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.legend(title='Bankrupt?', labels=['No', 'Yes'])
    plt.tight_layout()
    plt.show()
```
